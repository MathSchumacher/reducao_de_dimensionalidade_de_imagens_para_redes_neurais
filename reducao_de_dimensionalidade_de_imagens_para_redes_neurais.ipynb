```python
from PIL import Image
paisagem = Image.open('paisagem.png')
paisagem

def escala_cinza(imagem_colorida):
    w, h = imagem_colorida.size
    img = Image.new('RGB', (w, h))
    for x in range (w):
        for y in range(h):
          pxl = imagem_colorida.getpixel((x, y))
          # RGB multiplicado pelos pesos de iliminação para cada cor respectivamente
          lum = int(0.3 * pxl[0] + 0.59 * pxl[1] + 0.11 * pxl[2])
          img.putpixel((x, y), (lum, lum, lum))

    return img



def image_pb(escala_cinza):
    threshold = 128
    img_pb = escala_cinza.point(lambda p: 255 if p > threshold else 0)

    return img_pb



def img_resize(img):
    img = img.resize((555, 831))

    return img
#Função para redimencionar a imagem
paisagem = img_resize(paisagem)
paisagem

#Função para aplicar escalas de cinza à imagem
paisagem_gs = escala_cinza(paisagem)
paisagem_gs

#Função de converção para P&B
imagem_pb = image_pb(paisagem_gs)
imagem_pb


#Função para aplicar escalas de cinza à imagem
paisagem_gs = escala_cinza(paisagem)
paisagem_gs

#Função de converção para P&B
imagem_pb = image_pb(paisagem_gs)
imagem_pb
```
